```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
```

```python
def a_interval(sigma: float, gamma: float, n: int):    
    sample = np.random.normal(0, sigma, n)
    sum = np.sum(sample ** 2)
    right = sum / scipy.stats.chi2.ppf((1 - gamma) / 2, n)
    left = sum / scipy.stats.chi2.ppf((1 + gamma) / 2, n)
    return right - left
```

```python
def a_plot(sigma, gamma):
  x = np.array(range(10, 2000, 10))
  y = []
  for n in x:
    sum = 0
    for _ in range(0, 100):
        sum += a_interval(sigma, gamma, n)
    y.append(sum / 100)

  y = np.array(y)
  plt.grid()
  plt.plot(x, y)
  plt.xlabel('Number of samples')
  plt.ylabel('Interval length')
```

```python
a_plot(2, 0.1)
```

```python
def b_interval(sigma: float, gamma: float, n: int):    
    sample = np.random.normal(0, sigma, n)
    sum = np.mean(sample) ** 2
    right = n * sum / (scipy.stats.norm.ppf((3 - gamma) / 4) ** 2)
    left = n * sum / (scipy.stats.norm.ppf((3 + gamma) / 4) ** 2)
    return right - left
```

```python
def b_plot(sigma, gamma):
  x = np.array(range(10, 2000, 10))
  y = []
  for n in x:
    sum = 0
    for _ in range(0, 100):
        sum += b_interval(sigma, gamma, n)
    y.append(sum / 100)

  y = np.array(y)
  plt.grid()
  plt.plot(x, y)
  plt.xlabel('Number of samples')
  plt.ylabel('Interval length')
```

```python
b_plot(2, 0.1)
```
